```python
import numpy as np 
import pandas as pd
import statsmodels.formula.api as smf
from sklearn import preprocessing
```

```python
df = pd.read_excel("2023-07-27-GA4-&-sales-data-31-mar-to-20-Apr.xlsx")
print(df.head)
```

```python
# df = df[['Date','BVR Category Slug','State','Qty','Ordered Revenue','Commission','Traffic_landed on the category and product pages','Conversion','Average Bounce rate', 'Average Time spent on page', 'Phase']]
df.columns = ['date','category','state','qty','revenue','commission','traffic', 'clicks', 'ABR','time','phase']
# print(df.head)

# df_de = df.loc[(df['traffic'] > 0)]   # df_de is direct sale dataframe
# df_ide = df.loc[(df['traffic'] == 0)]   # df_ide is Indirect sale dataframe

# print(df_ds.head)
# print(df_ide.head)

# df_price = df.loc[(df['phase']==1)]
# df_wo_price = df.loc[(df['phase']==0)]

state_df = df.drop(['category'], axis=1)
state_df = state_df[state_df.state != 'Rest of US']
# state_df = state_df[(state_df.date != '2023-04-04') & (state_df.date != '2023-04-06') & (state_df.date != '2023-04-11') & (state_df.date != '2023-04-13')]

# print(state_df.head)

master_df = state_df.groupby(['date','state','phase'], as_index=False).agg({'qty':'sum', 'revenue':'sum', 'commission':'sum', 'traffic': 'sum', 'clicks':'sum', 'ABR':'mean', 'time':'mean'})
master_df['orderperclick'] = master_df['qty']/master_df['clicks']
master_df['itemprice'] = master_df['revenue']/master_df['qty']
master_df['logT'] = np.log10(master_df['time']+1)
# state_df = state_df[(state_df.date != '2023-04-04') & (state_df.date != '2023-04-06') & (state_df.date != '2023-04-11') & (state_df.date != '2023-04-13')]

big_state_df = master_df[(master_df.state =='California')].reset_index()


label_encoder = preprocessing.LabelEncoder()

master_df['encodedstate']= label_encoder.fit_transform(master_df['state'])

# print(master_df['encodedstate'].unique())
  
print(big_state_df.head)


# print(master_df.head)

# click_df = state_df.groupby(['date','state','phase'])['clicks'].sum().reset_index()

# click_df_price = master_df.loc[(master_df['phase']==1)]
# click_df_wo_price = master_df.loc[(master_df['phase']==0)]
# mean_click_df_price= click_df_price['clicks'].mean()
# mean_click_df_wo_price = click_df_wo_price['clicks'].mean()

# print("----------------------------")
# print("mean_click_df_price==== ",master_df.loc[(master_df['phase']==1)]['clicks'].mean())
# print("mean_click_df_wo_price==== ",master_df.loc[(master_df['phase']==0)]['clicks'].mean())
# print("----------------------------")
# print("orderperclick price==== ",master_df.loc[(master_df['phase']==1)]['orderperclick'].mean())
# print("orderperclick wo_price==== ",master_df.loc[(master_df['phase']==0)]['orderperclick'].mean())
# print("----------------------------")
# print("qty price==== ",master_df.loc[(master_df['phase']==1)]['qty'].mean())
# print("qty wo_price==== ",master_df.loc[(master_df['phase']==0)]['qty'].mean())
# print("----------------------------")
# print("revenue price==== ",master_df.loc[(master_df['phase']==1)]['revenue'].mean())
# print("revenue wo_price==== ",master_df.loc[(master_df['phase']==0)]['revenue'].mean())
# print("----------------------------")
# print("itemprice price==== ",master_df.loc[(master_df['phase']==1)]['itemprice'].mean())
# print("itemprice wo_price==== ",master_df.loc[(master_df['phase']==0)]['itemprice'].mean())
# print("----------------------------")
# print("logT price==== ",master_df.loc[(master_df['phase']==1)]['logT'].mean())
# print("logT wo_price==== ",master_df.loc[(master_df['phase']==0)]['logT'].mean())
# print("----------------------------")
# print("ABR price==== ",master_df.loc[(master_df['phase']==1)]['ABR'].mean())
# print("ABR wo_price==== ",master_df.loc[(master_df['phase']==0)]['ABR'].mean())




# state_df_price = df_price.groupby(['state','category'])
# state_df_price.first()

# print(click_df.head)

```

```python
# Dependent Variables: Clicks 

clicks_model = smf.ols(formula='clicks ~ phase', data=master_df).fit(cov_type='HC3') 
print(clicks_model.summary())
```

```python
# Dependent Variables: orderperclick 

orderperclick_model = smf.ols(formula='orderperclick ~ phase', data=master_df).fit(cov_type='HC3') 
print(orderperclick_model.summary())
```

```python
# Dependent Variables: no. of orders

orders_model = smf.ols(formula='qty ~ phase', data=master_df).fit(cov_type='HC3') 
print(orders_model.summary())
```

```python
# Dependent Variables: Dollar value of items - revenue

dollar_model = smf.ols(formula='revenue ~ phase', data=master_df).fit(cov_type='HC3') 
print(dollar_model.summary())
```

```python
# Dependent Variables: Dollar value of items - revenue

itemprice_model = smf.ols(formula='itemprice ~ phase', data=master_df).fit(cov_type='HC3') 
print(itemprice_model.summary())
```

```python
# Dependent Variables: average time spent - log(T+1)

logtime_model = smf.ols(formula='logT ~ phase', data=master_df).fit(cov_type='HC3') 
print(logtime_model.summary())
```

```python
# Dependent Variables: average time spent - log(T+1)

ABR_model = smf.ols(formula='ABR ~ phase', data=master_df).fit(cov_type='HC3') 
print(ABR_model.summary())
```

```python
# Dependent Variables: Clicks for california

big_state_df_clicks_model = smf.ols(formula='clicks ~ phase', data=big_state_df).fit(cov_type='HC3') 
print(big_state_df_clicks_model.summary())
```
